## Часть 4: Задания для практики
Прочитайте данные из файла 'titanic.csv'.
Проверьте, есть ли пропущенные значения в данных и заполните их нулями.
Выведите первые 10 строк данных.
Выберите только те строки, где значение в столбце 'Age' больше 30.
Отсортируйте данные по столбцу 'Fare' в порядке убывания.
Сгруппируйте данные по столбцу 'Pclass' и вычислите средний возраст ('Age') для каждого класса.

```python
# Проверка на NaN и заполнение их нулями
data.isna()
filled_data = data.fillna(0)

# первыe 10 строк
filled_data.head(10)

#'Age' больше 30
age_above_30 = filled_data[filled_data['Age'] > 30]
age_above_30

# Сортировка данных по Fare в порядке убывания
sorted_by_fare = filled_data.sort_values(by='Fare', ascending=False)
sorted_by_fare

# Группировка по 'Pclass' и вычисление среднего возраста
average_age_by_pclass = filled_data.groupby('Pclass')['Age'].mean()
average_age_by_pclass
```
